```python
df_list = []
for month in ["09", "10", "11"]:
    df = pd.read_csv(
        f"../level1d/header-2020-{month}-release_8.0-000000_POLARSTERN.psv", 
        delimiter="|", 
        header=None,
        names=names,
        index_col="report_id",
    )
    df_list.append(df)

data = pd.concat(df_list, axis=0)
data
```

```python
position_check = marine_qc.do_position_check(
    lat=data["latitude"],
    lon=data["longitude"],
)
position_check[position_check == 1]
```

```python
date_check = marine_qc.do_date_check(
    date=data["report_timestamp"],
)
date_check[date_check == 1]
```

```python
time_check = marine_qc.do_time_check(
    date=data["report_timestamp"],
)
time_check[time_check == 1]
```

```python
track_flags = marine_qc.do_track_check(
    vsi=data["station_speed"],
    dsi=data["station_course"],
    lat=data["latitude"],
    lon=data["longitude"],
    date=data["report_timestamp"],
    max_direction_change=60.0,
    max_speed_change=10.0,
    max_absolute_speed=40.0,
    max_midpoint_discrepancy=150.0,
)
track_flags[track_flags == 1]
```

```python
iquam_flags = marine_qc.do_iquam_track_check(
    lat=data["latitude"],
    lon=data["longitude"],
    date=data["report_timestamp"],
    speed_limit=60.0,
    delta_d=1.11,
    delta_t=0.01,
    n_neighbours=5,
)
iquam_flags[iquam_flags == 1]
```
